## Creating dataframe for conserning values

```python
cols=['Ticker', 'Price','One-Year Price Return', 'Number Of Shares to Buy']
stock_param=pd.DataFrame(columns=cols )
```

Creating batches for faster api requests

```python
batch_req=[]
res=list(create_batches(sp500['Ticker'],100))
for i in res:
    batch_req.append(','.join(i))
print(batch_req)
```

```python
#updated way
for symbol in batch_req:
    ot=requests.get(f"https://api.iex.cloud/v1/data/CORE/ADVANCED_STATS/{symbol}?token={key_iex}").json()
    sym=symbol.split(',')
    #print(sym)
    for i in range(0,len(ot)):
        cs=ot[i]
        stock_param.loc[len(stock_param.index)] = [sym[i], 
                                                    stock_prices.loc[len(stock_param.index),'Price'], 
                                                    cs["year1ChangePercent"], 
                                                    'N/A']
```

```python
stock_param
```

Sorting these values according to yearlly return to find to 50 with best return

```python
stock_param.sort_values('One-Year Price Return',ascending=False,inplace=True)
```

```python

super50=stock_param[:50]

super50
```

```python
super50.reset_index(inplace=True)
```

```python
super50=super50.drop('index',axis=1)
```

```python
super50
```

Now lets distribute the money in these stocks according to the  return ratio

```python
su=super50['One-Year Price Return'].sum()
print(su)
```

Lets invest total value of 10000000

```python
super50.info()
super50['One-Year Price Return']=pd.to_numeric(super50['One-Year Price Return'])
super50.info()
```

```python
for i in range (0,50):
    super50.loc[i,'Number Of Shares to Buy']=math.floor(((super50.loc[i,'One-Year Price Return']/su)*10000000)/ super50.loc[i,'Price'])
super50
```
