Now instead of checking an early return we check the return on other intervals like 3 onths,6 months to check how stable and consistent actually have the returns been

```python
hqm_columns = [
                'Ticker', 
                'Price', 
                'Number of Shares to Buy', 
                'One-Year Price Return', 
                'One-Year Return Percentile',
                'Six-Month Price Return',
                'Six-Month Return Percentile',
                'Three-Month Price Return',
                'Three-Month Return Percentile',
                'One-Month Price Return',
                'One-Month Return Percentile',
                'HQM Score'
                ]

hqm_param = pd.DataFrame(columns = hqm_columns)
```

```python
for symbol in batch_req:
    ot=requests.get(f"https://api.iex.cloud/v1/data/CORE/ADVANCED_STATS/{symbol}?token={key_iex}").json()
    sym=symbol.split(',')
    #print(sym)
    for i in range(0,len(ot)):
        cs=ot[i]
        hqm_param.loc[len(hqm_param.index)] = [sym[i], 
                                                    stock_prices.loc[len(hqm_param.index),'Price'], 
                                                    5,
                                                    cs["year1ChangePercent"], 
                                                     5,
                                                    cs["month6ChangePercent"], 
                                                    5,
                                                    cs["month3ChangePercent"], 
                                                     5,
                                                    cs["month1ChangePercent"], 
                                                     5,
                                                    5]

```

```python
hqm_param
```

```python
hqm_param.dropna(axis=0,inplace=True)
```

```python
hqm_param.info()
hqm_param.reset_index(inplace=True)
hqm_param
```

```python
hqm_param.drop('index',axis=1,inplace=True)
hqm_param
```

```python
time_periods = [
                'One-Year',
                'Six-Month',
                'Three-Month',
                'One-Month'
                ]
for j in time_periods:
        hqm_param[f'{j} Price Return']=pd.to_numeric(hqm_param[f'{j} Price Return'])
for i in range (0,len(hqm_param.index)):
    for j in time_periods:
        hqm_param.loc[i,f'{j} Return Percentile']=stats.percentileofscore(hqm_param[f'{j} Price Return'],hqm_param.loc[i,f'{j} Price Return'])/100
#hqm_param
```

```python
hqm_param
```

Now we will calculate hqm score as average of each time period return percentile

```python

tem=[]
for j in time_periods:
    tem.append(f'{j} Return Percentile')
for i in range(0,len(hqm_param.index)):
    su=[]
    for j in tem:
        su.append(hqm_param.loc[i,j])
    hqm_param.loc[i,"HQM Score"]=np.mean(su)
   
```

```python
hqm_param
```

```python
hqm_param.sort_values('HQM Score',ascending=False,inplace=True)
hqm_param.reset_index(inplace=True)
hqm_param.drop('index',axis=1,inplace=True)
```

```python
best50=hqm_param[:50]
best50
```

Now lets format it to save in excel

```python
writer = pd.ExcelWriter('momentum_strategy.xlsx', engine='xlsxwriter')
hqm_param.to_excel(writer, sheet_name='Momentum Strategy', index = False)
```

```python

dollar_template = writer.book.add_format(
        {
            'num_format':'$0.00',
            
        }
    )

integer_template = writer.book.add_format(
        {
            'num_format':'0',
            
        }
    )

percent_template = writer.book.add_format(
        {
            'num_format':'0.0%',
            
        }
    )
float_template = writer.book.add_format(
        {
            'num_format':'0.0000',
            
        }
    )
```

```python
column_formats = { 
                    
                    'B': ['Price', dollar_template],
                    'C': ['Number of Shares to Buy', integer_template],
                    'D': ['One-Year Price Return', percent_template],
                    'E': ['One-Year Return Percentile', percent_template],
                    'F': ['Six-Month Price Return', percent_template],
                    'G': ['Six-Month Return Percentile', percent_template],
                    'H': ['Three-Month Price Return', percent_template],
                    'I': ['Three-Month Return Percentile', percent_template],
                    'J': ['One-Month Price Return', percent_template],
                    'K': ['One-Month Return Percentile', percent_template],
                    'L': ['HQM Score', float_template]
                    }

for column in column_formats.keys():
    writer.sheets['Momentum Strategy'].set_column(f'{column}:{column}', 20, column_formats[column][1])
    writer.sheets['Momentum Strategy'].write(f'{column}1', column_formats[column][0])
```
